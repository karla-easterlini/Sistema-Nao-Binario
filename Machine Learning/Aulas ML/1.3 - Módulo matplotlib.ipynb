Guia rápido do módulo matplotlib
================================



O módulo `matplotlib` é utilizado para a construção de gráficos (usualmente chamados de plots). Para isso, importamos o módulo `pyplot` dentro do `matplotlib` com o apelido de `plt`. Vamos importar também o `numpy` pois ele será útil para gerar dados para nossos gráficos.



```python
from matplotlib import pyplot as plt
import numpy as np
```

## Criando um gráfico de linhas



O `matplotlib` tem dois modos de operação. Um deles imita a forma como o `MATLAB` produz gráficos. Aqui nós *não* veremos esta forma de plotar gráficos, mas sim veremos a forma baseada em objetos.

Primeiro criamos os objetos de `figura` e `eixo` com a função `plt.subplots`.



```python
figura, eixo = plt.subplots()
```

O gráfico está vazio! Antes de seguir, precisamos de dados para plotar. Vamos plotar a função seno. Para plotar, usamos nosso objeto `eixo`. Para mostrar sua figura em um notebook Jupyter, basta que a última linha seja o nome do seu objeto de figura.



```python
pi = 3.1415
x = np.linspace(0, 2 * pi)

eixo.plot(x, np.sin(x))

figura
```

Podemos acresentar outras linhas no nosso gráfico!



```python
eixo.plot(x, np.cos(x))  # cosseno
eixo.plot(x, x)          # identidade
figura
```

## Colocando legenda no gráfico



Para mostrar uma legenda no seu gráfico, use o `eixo.legend`. Observe que para isso funcionar corretamente você precisa dar o nome da legenda no argumento `label` quando for usar o `eixo.plot`.



```python
figura, eixo = plt.subplots()

pi = 3.1415
x = np.linspace(0, 2 * pi)

eixo.plot(x, np.sin(x), label="Seno")
eixo.plot(x, np.cos(x), label="Cosseno")
eixo.plot(x, x, label="Identidade")

eixo.legend()
```

Você pode controlar a posição da legenda!



```python
eixo.legend(loc=1)
figura
```

```python
eixo.legend(loc=2)
figura
```

```python
eixo.legend(loc=3)
figura
```

```python
eixo.legend(loc=4)
figura
```

## Criando um gráfico de pontos



Para criar um gráfico de pontos, você ainda irá usar o `eixo.plot`, porém deverá passar um valor para o símbolo que será utilizado para seus pontos (argumento `marker`) e deverá passar a string `"none"` para o argumento `linestyle` para que o matplotlib não plote a linha.



```python
figura, eixo = plt.subplots()

x = np.linspace(0, 10)
y = x**2

eixo.plot(
    x,
    y,
    marker="o",
    linestyle="none",
    label="Meus dados",
)

eixo.legend()
```

Nada impede que você misture dois tipos de plots no mesmo gráfico!



```python
eixo.plot(x, np.sin(x), label="Seno")

eixo.legend()
figura
```

## Nomes dos eixos e título do gráfico



É possível acrescertar nomes aos eixos e dar um título para seu gráfico.



```python
figura, eixo = plt.subplots()

pi = 3.1415
x = np.linspace(0, 2 * pi)

eixo.plot(x, np.sin(x), label="Seno")
eixo.plot(x, np.cos(x), label="Cosseno")

eixo.legend()

eixo.set_title("Funções trigonométricas")
eixo.set_xlabel("Ângulo")
eixo.set_ylabel("Valor")
```

## Escala logarítmica



Você pode alterar tanto o eixo $x$ quanto o eixo $y$ para ser exibido em escala logarítmica. Veja os exemplos abaixo



```python
figura, eixo = plt.subplots()

pi = 3.1415
x = np.linspace(0, 10 * pi, 500)

eixo.plot(x, np.sin(x), label="Seno")
eixo.plot(x, np.cos(x), label="Cosseno")

eixo.legend()

eixo.set_title("Funções trigonométricas")
eixo.set_xlabel("Ângulo")
eixo.set_ylabel("Valor")

eixo.set_yscale('log', nonpositive='clip')
```

## Criando um histograma



Histogramas são gráficos que nos mostram uma representação da distribuição de um conjunto de dados numéricos. Histogramas separam os dados em compartimentos (ou caixas) no eixo $x$ e representam sua frequência ou quantidade no eixo $y$.

Antes de criar um histograma, precisamos dos dados! Vamos sortear alguns dados aleatórios com o `numpy`.



```python
meus_dados = np.random.random(50)
```

Para plotar um histograma, use o método `hist` do eixo.



```python
figura, eixo = plt.subplots()
eixo.hist(meus_dados, label="Distribuição dados aleatórios")
eixo.set_xlabel("Variável")
eixo.set_ylabel("Quantidade")
eixo.legend()
```

Você pode alterar o número de caixas do seu histograma!



```python
figura, eixo = plt.subplots()
num_caixas = 20
eixo.hist(meus_dados, bins=num_caixas, label="Distribuição dados aleatórios")
eixo.set_xlabel("Variável")
eixo.set_ylabel("Quantidade")
eixo.legend()
```

## Linhas verticais e horizontais



Para plotar linhas verticais ou horizontais use os métodos `axvline` e `axhline`.



```python
figura, eixo = plt.subplots()

quantidade, caixas, _ = eixo.hist(
    meus_dados, label="Distribuição dados aleatórios"
)

valor_medio = np.mean(meus_dados)
eixo.axvline(valor_medio, linestyle="--", color="red", label="Valor médio")

quantidade_maxima = max(quantidade)
eixo.axhline(
    quantidade_maxima, linestyle="--", color="black", label="Quantidade máxima"
)

eixo.set_xlabel("Variável")
eixo.set_ylabel("Quantidade")
eixo.legend(loc=4)
```

## Criando um boxplot



```python
dados = []

# gerando dados sintéticos
for desvio_padrao in range(1, 4):
    dados.append(np.random.normal(0, desvio_padrao, size=100))

rotulos = ["Dados 1", "Dados 2", "Dados 3"]

figura, eixo = plt.subplots()

eixo.boxplot(dados, labels=rotulos)

eixo.set_ylabel("Valores")

```
