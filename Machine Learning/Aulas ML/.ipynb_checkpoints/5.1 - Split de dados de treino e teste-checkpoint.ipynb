Avaliação de modelos: a estratégia de divisão entre treino e teste
==================================================================



## Pergunta



Suponha que você fez todo o processo de coletar dados e treinar um modelo de aprendizado de máquina. Com seu modelo em mãos você pode usar ele para realizar previsões!

Quando realizamos previsões, usualmente desejamos prever o nosso target para dados *diferentes* dos dados utilizados para o treinamento do modelo (afinal, os dados utilizados para o treinamento nós já conhecemos o valor do target, não é?).

A pergunta é: **como podemos avaliar a performance de modelos para dados fora do conjunto de dados utilizado para o treinamento do modelo?**



## Perfeitamente equilibrado, como todas as coisas deveriam ser



Os gráficos gerados do código abaixo nos apresentam 3 modelos diferentes (curva vermelha). Estes modelos se propõe a ajustar os dados (pontos azuis). Qual destes 3 modelos você escolheria para modelar seus dados?



```python
# Código modificado de https://scikit-learn.org/stable/auto_examples/model_selection/plot_underfitting_overfitting.html

import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

np.random.seed(0)  # fixando a semente para os sorteios aleatórios

NUM_PONTOS = 30
COMPLEXIDADE_DOS_MODELOS = [1, 4, 15]


def funcao_verdadeira(X):
    """Esta é a nossa função verdadeira que queremos descobrir/modelar."""
    return np.cos(1.5 * np.pi * X)


# criando os dados sintéticos
X = np.sort(np.random.rand(NUM_PONTOS))
y = funcao_verdadeira(X) + np.random.randn(NUM_PONTOS) * 0.1

X_continuo = np.linspace(0, 1, 100)

fig, axes = plt.subplots(
    ncols=3,
    nrows=1,
    figsize=(len(COMPLEXIDADE_DOS_MODELOS) * 5, 5),
    sharey=True,
)

for i in range(len(COMPLEXIDADE_DOS_MODELOS)):

    eixo = axes[i]

    # Este bloco cria o pipeline do modelo que será ajustado
    polynomial_features = PolynomialFeatures(
        degree=COMPLEXIDADE_DOS_MODELOS[i], include_bias=False
    )
    linear_regression = LinearRegression()
    pipeline = Pipeline(
        [
            ("features_polinomiais", polynomial_features),
            ("regressao_linear", linear_regression),
        ]
    )
    pipeline.fit(X[:, np.newaxis], y)

    # Plotando!
    eixo.plot(
        X_continuo,
        pipeline.predict(X_continuo[:, np.newaxis]),
        label="Modelo",
        color="r",
    )
    eixo.scatter(X, y, edgecolor="b", s=20, label="Dados")

    # Configurando o plot
    eixo.set_xlabel("x")
    eixo.set_ylabel("y")
    eixo.set_xlim((0, 1))
    eixo.set_ylim((-2, 2))
    eixo.legend()

plt.show()
```

Antes de seguir em frente, é importantíssimo que esteja claro o que é considerado um *bom modelo*. Um bom modelo é aquele que tem um balanço aceitável entre viés e variância e por conta disso é capaz de prever suficientemente bem o *target* de interesse.

Um modelo com alta variância e baixo viés está <u>sobreajustado</u> aos dados (também conhecido como *overfitting*). Ele descreve muito bem os dados que conhece, porém erra muito quando tenta prever dados que nunca viu. No jargão informal da área, dizemos que o modelo &ldquo;bitolou&rdquo; nos dados apresentados.

Um modelo com baixa variância e alto viés está <u>subajustado</u> aos dados (também conhecido como *underfitting*). Ele não consegue descrever bem os dados que conhece e também erra muito quando tenta prever dados que nunca viu. Um modelo subajustado simplesmente não consegue capturar a complexidade do fenômeno sendo modelado.

Um modelo com um bom equilíbrio entre viés e variância é um modelo que consegue descrever razoavelmente bem os dados que conhece e também acerta razoavelmente bem quando tenta prever dados que nunca viu. É este o tipo de modelo que consideramos um *bom modelo*!

Agora que você sabe sobre viés e variância, você consegue identificar nos 3 gráficos apresentados no início desta seção quais são os 3 diferentes tipos de modelos discutidos acima?

Não deixe de ver esse [excelente vídeo](https://www.youtube.com/watch?v=EuBBz3bI-aA) do StatQuest para visualizar o conceito acima e ler este [post](https://mineetha.com/2021/02/25/bias-variance-tradeoff/) da Mineetha Chandralekha para sedimentar seu conhecimento.



## Estratégia de holdout



Observe que usamos os termos &ldquo;dados que o modelo conhece&rdquo; e &ldquo;dados que o modelo não conhece&rdquo; na seção acima. Vamos defini-los:

-   *Dados que o modelo conhece* são os dados que o modelo teve acesso (direta ou indiretamente!) durante seu processo de treino. O processo de treino de um modelo é o momento em que ele é capaz de alterar seus parâmetros internos.

-   *Dados que o modelo não conhece* é o conjunto de todos os dados possíveis/existentes menos os dados que o modelo conhece. Como estes dados não estavam disponíveis durante o processo de treino do modelo, eles não influenciaram os parâmetros internos do modelo.

Tendo em vista tudo que foi discutido, vamos finalmente ver uma estratégia para tentar obter um bom modelo usando aprendizado de máquina. Trata-se da <u>estratégia de holdout</u>, também conhecida como <u>divisão em treino e teste</u> ou <u>train test split</u>.

Esta estratégia é relativamente simples, ela consiste em reservar uma fração dos dados que temos e usar apenas o restante dos dados para o treino do modelo. Após o treino, nós tentamos prever os dados que reservamos e, com isso, checamos se nosso modelo tem uma performance razoável quando tenta prever dados que não conhece.

Ao conjunto de dados reservados, que não serão utilizados para o treino do modelo, damos o nome de dados de *teste* ou dados *holdout*.

Ao conjunto de dados que usamos para treinar o modelo damos o nome de dados de *treino*.



## A função `train_test_split`



Para dividir um conjunto de dados em treino e teste temos a função `train_test_split` do módulo `scikit-learn`.

Aqui nós veremos como usar essa função em conjunto com um `DataFrame` de `pandas`. No entanto, usualmente na documentação oficial e nos tutoriais online você verá esta função sendo usada com arrays de `numpy`.

Vamos primeiramente criar nosso `DataFrame`. Aqui usaremos conjunto de dados sobre pinguins disponível no `seaborn`.



```python
import seaborn as sns

DATASET_NAME = "penguins"
FEATURES = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm"]
TARGET = ["body_mass_g"]

df = sns.load_dataset(DATASET_NAME)

df = df.reindex(FEATURES + TARGET, axis=1)
df = df.dropna()  # remove linhas com células vazia

print(df)
```

Para usar a função `train_test_split`, precisamos definir quanto dos dados vamos reservar para o conjunto de teste. Aqui podemos escolher uma porcentagem (isso vai depender de quantos dados você tem, um valor entre 10% a 33% é o que usualmente se encontra na literatura) ou um número absoluto de exemplos que irão compor o conjunto de teste. Essa é uma escolha importante! Se tivermos poucos dados no conjunto de teste não teremos muita confiança na performance do nosso modelo (afinal, ele pode se sair bem por pura sorte!). Se tivermos muitos dados no conjunto de teste, nós &ldquo;perdemos&rdquo; dados que poderíamos usar para treinar nosso modelo (o que deixará mais difícil o processo de encontrar um bom modelo).

Para este exemplo vamos reservar 10% dos dados para o conjunto de teste. Para fazer isso, iremos passar o valor `0.1` para o argumento `test_size`. Se quiséssemos 20% deveríamos passar o valor `0.2` para este argumento. Se quiséssemos reservar exatamente 150 exemplos, deveríamos passar o valor `150` para este argumento. Números menores que 1 representam porcentagem e número maiores ou iguais a 1 representam quantidade absoluta de exemplos.



```python
from sklearn.model_selection import train_test_split

TAMANHO_TESTE = 0.1

indices = df.index
indices_treino, indices_teste = train_test_split(
    indices, test_size=TAMANHO_TESTE
)

df_treino = df.loc[indices_treino]
df_teste = df.loc[indices_teste]

print(df_treino)
print()
print(df_teste)
```

No código acima nós usamos a função `train_test_split` para separar os *índices* do `DataFrame`. Lembre-se que os índices são <u>identificadores únicos</u> de cada exemplo do nosso conjunto de dados, logo podemos recuperar esses exemplos usando o método `loc` do `DataFrame`.

Cada vez que você rodar o código acima você irá obter um conjunto de teste e treino diferentes. Isso ocorre pois a função `train_test_split` faz um sorteio (pseudo)-aleatório dos dados que recebe.

Em geral, quando estamos tratando da estratégia de *holdout*, nós queremos que o conjunto de dados reservados seja sorteado aleatoriamente, porém que seja *sempre o mesmo*. Para resolver isso, basta passar um número inteiro positivo qualquer para o argumento `random_state` e você terá garantia de que seu conjunto de treino e teste serão sempre os mesmos.



```python
TAMANHO_TESTE = 0.1
SEMENTE_ALEATORIA = 61455

indices = df.index
indices_treino, indices_teste = train_test_split(
    indices, test_size=TAMANHO_TESTE, random_state=SEMENTE_ALEATORIA
)

df_treino = df.loc[indices_treino]
df_teste = df.loc[indices_teste]

print(df_treino)
print()
print(df_teste)
```

Pode rodar a célula anterior quantas vezes quiser, ela irá sempre retornar o mesmo conjunto de treino e teste!



## Avaliando um modelo



Vamos usar os dados acima para treinar um modelo $k$​-NN seguindo a estratégia de holdout. Primeiramente precisamos dos atributos e targets do treino e do teste.



```python
X_treino = df_treino.reindex(FEATURES, axis=1).values
y_treino = df_treino.reindex(TARGET, axis=1).values.ravel()

X_teste = df_teste.reindex(FEATURES, axis=1).values
y_teste = df_teste.reindex(TARGET, axis=1).values.ravel()
```

Agora vamos criar a instância do modelo.



```python
from sklearn.neighbors import KNeighborsRegressor

num_vizinhos = 3
modelo_knn = KNeighborsRegressor(n_neighbors=num_vizinhos)
```

Segundo a estratégia de holdout, o modelo só pode ser treinado com os dados de treino, **não podemos apresentar os dados de teste nesta etapa**!



```python
modelo_knn.fit(X_treino, y_treino)
```

Agora que o modelo está treinado, podemos observar sua performance no conjunto de dados de teste.



```python
from sklearn.metrics import mean_squared_error

X_verdadeiro = X_teste
y_verdadeiro = y_teste

y_previsto = modelo_knn.predict(X_verdadeiro)

RMSE = mean_squared_error(y_verdadeiro, y_previsto, squared=False)

print(f"O RMSE do modelo foi de {RMSE} unidades de y ao quadrado.")
```

Observe que este valor é maior do que o valor do RMSE que obtivemos quando *treinamos* o modelo $k$-NN com todos os dados e *testamos* o modelo com os mesmos dados usados para o treino. (Lembre-se que quanto *menor* o RMSE *melhor* a performance do modelo).

Quando dados de treino são usados para testar a performance de modelos, nós *superestimamos* a performance do modelo! Isso pode parecer interessante, mas não é! É muito ruim não saber a real performance dos modelos que treinamos.
