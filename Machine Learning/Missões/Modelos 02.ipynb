### Loadando o dataset e fazendo o split

```python
import seaborn as sns
data_set = "mpg"
df = sns.load_dataset(data_set)
df
```

O dadaset escolhido nesse caso foi o que contém informações sobre o rendimento de carros (miles per galon, mpg), também vindo do módulo *_Seaborn_*

```python
from sklearn import neighbors
from sklearn.model_selection import train_test_split

FEATURES = ["cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year"]
TARGET = ["mpg"]

df.dropna(inplace=True)

df = df.reindex(FEATURES + TARGET, axis=1)

test_size = 0.1
random_seed = 69420

index = df.index
train_index, test_index = train_test_split(index, test_size=test_size, random_state = random_seed)

df_train = df.loc[train_index]
df_test = df.loc[test_index]

print(f'Dados treino: \n {df_train}')
print(90*'-')
print(f'Dados Teste: \n {df_test}')
```

### É hora do Show!! (Vamos treinar o modelo.)

```python
X_treino = df_train.reindex(FEATURES, axis=1).values
y_treino = df_train.reindex(TARGET, axis=1).values.ravel()

X_teste = df_test.reindex(FEATURES, axis=1).values
y_teste = df_test.reindex(TARGET, axis=1).values.ravel()
```

```python
from sklearn.neighbors import KNeighborsRegressor

vizin = 3
modelo_knn_3_vizin = KNeighborsRegressor(n_neighbors=vizin)
modelo_knn_3_vizin.fit(X_treino, y_treino)
```

#### Show feito, hora de ver se deu certo!

É um RMSE relativamente bom, mas como será que a performance do nosso modelo se comporta quando aumentamos a quantidade de vizinhos analisado? Essa é pergunta que iremos analisar e responder nessa próxima sessão!
