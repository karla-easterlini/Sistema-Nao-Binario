O andar do bêbado
=================



## Objetivo



Um senhor que tomou uma ou duas doses a mais do que deveria está andando por um campo plano, infinito e livre de obstáculos. A cada minuto, este senhor dá um passo de um metro para norte, sul, leste ou oeste. É sempre um passo por minuto e sempre em apenas uma direção (todas as direções têm a mesma chance de serem escolhidas).

Faça gráficos mostrando:

1.  A trajetória do senhor andante no plano;

2.  A posição em $x$ do senhor andante em função do tempo (plote uma linha indicando o valor médio);

3.  A posição em $y$ do senhor andante em função do tempo (plote uma linha indicando o valor médio);

4.  A distância relativa percorrida pelo senhor andante em função do tempo (a distância relativa é a distância Euclidiana entre a posição no tempo atual e a posição no tempo zero).

Considere que o senhor andante inicia sua trajetória no tempo $t=0$ na coordenada $(0,0)$. Considere um tempo de pelo menos 1000 minutos para plotar seus gráficos. Discuta sobre os resultados obtidos.



## Resultado e discussão



```python
import matplotlib.pyplot as plt
import random

tempo_escolhido = 1000

instrucoes = {
    "Norte": [0, 1],
    "Sul": [0, -1],
    "Leste": [1, 0],
    "Oeste": [-1, 0]
}

direcoes = list(instrucoes.keys())

posicao_atual = [0, 0] 
passos_em_x = []
passos_em_y = []

def escolhe_direcao(dic):
    escolha = random.choice(dic)
    return escolha

for i in range(tempo_escolhido):
    escolha = escolhe_direcao(direcoes)
    movimento = instrucoes[escolha]
    
    posicao_atual[0] += movimento[0]  # Muda em x
    posicao_atual[1] += movimento[1]  # muda em y
    
    passos_em_x.append(posicao_atual[0])
    passos_em_y.append(posicao_atual[1])


somax = 0    
for x in passos_em_x:
    somax += passos_em_x[x]
mediax = somax/len(passos_em_x)


somay = 0    
for y in passos_em_y:
    somay += passos_em_y[y]
mediay = somay/len(passos_em_y)

    
    


plt.figure(figsize=(10, 8))
plt.plot(passos_em_x, passos_em_y, label='Trajetória')
plt.xlabel('Posição em X')
plt.ylabel('Posição em Y')
plt.title('Trajetória do bêbado em x e y')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(range(tempo_escolhido), passos_em_x, label='Trajetória')
plt.axhline(mediax, color='red', linestyle='--', label='Valor Médio')
plt.xlabel('Posição em X')
plt.ylabel('Posição em Y')
plt.title('Trajetória do bêbado em x')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(range(tempo_escolhido), passos_em_y, label='Trajetória')
plt.axhline(mediay, color='green', linestyle='--', label='Valor Médio')
plt.xlabel('Posição em X')
plt.ylabel('Posição em Y')
plt.title('Trajetória do bêbado em y')
plt.legend()
plt.grid(True)
plt.show()

```
