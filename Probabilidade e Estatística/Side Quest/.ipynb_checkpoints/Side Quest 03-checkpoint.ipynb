O andar do bêbado
=================



## Objetivo



Um senhor que tomou uma ou duas doses a mais do que deveria está andando por um campo plano, infinito e livre de obstáculos. A cada minuto, este senhor dá um passo de um metro para norte, sul, leste ou oeste. É sempre um passo por minuto e sempre em apenas uma direção (todas as direções têm a mesma chance de serem escolhidas).

Faça gráficos mostrando:

1.  A trajetória do senhor andante no plano;

2.  A posição em $x$ do senhor andante em função do tempo (plote uma linha indicando o valor médio);

3.  A posição em $y$ do senhor andante em função do tempo (plote uma linha indicando o valor médio);

4.  A distância relativa percorrida pelo senhor andante em função do tempo (a distância relativa é a distância Euclidiana entre a posição no tempo atual e a posição no tempo zero).

Considere que o senhor andante inicia sua trajetória no tempo $t=0$ na coordenada $(0,0)$. Considere um tempo de pelo menos 1000 minutos para plotar seus gráficos. Discuta sobre os resultados obtidos.



## Resultado e discussão



```python
import matplotlib.pyplot as plt
import random

tempo_escolhido = 1000

instrucoes = {
    "Norte": [0, 1],
    "Sul": [0, -1],
    "Leste": [1, 0],
    "Oeste": [-1, 0]
}

direcoes = list(instrucoes.keys())

posicao_atual = [0, 0] 
posicao_inicial = [0, 0]
passos_em_x = []
passos_em_y = []
dist_relativa = []

def dist_euclid(x1, y1, x2, y2):
    return (((x1 - x2) ** 2 + (y1 - y2) ** 2)**(1/2))

def escolhe_direcao(dic):
    escolha = random.choice(dic)
    return escolha

for i in range(tempo_escolhido):
    escolha = escolhe_direcao(direcoes)
    movimento = instrucoes[escolha]
    
    posicao_atual[0] += movimento[0]  # Muda em x
    posicao_atual[1] += movimento[1]  # muda em y
    
    passos_em_x.append(posicao_atual[0])
    passos_em_y.append(posicao_atual[1])
    
    distancia1 = dist_euclid(posicao_atual[0], posicao_atual[1], posicao_inicial[0], posicao_inicial[1])
    dist_relativa.append(distancia1)


somax = 0    
for x in passos_em_x:
    somax += passos_em_x[x]
mediax = somax/len(passos_em_x)


somay = 0    
for y in passos_em_y:
    somay += passos_em_y[y]
mediay = somay/len(passos_em_y)


plt.figure(figsize=(10, 8))
plt.plot(passos_em_x, passos_em_y, label='Trajetória')
plt.xlabel('Posição em X')
plt.ylabel('Posição em Y')
plt.title('Trajetória do bêbado em x e y')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(range(tempo_escolhido), passos_em_x, label='Trajetória')
plt.axhline(mediax, color='red', linestyle='--', label='Valor Médio')
plt.xlabel('Tempo')
plt.ylabel('Posição em X')
plt.title('Trajetória do bêbado em x')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(range(tempo_escolhido), passos_em_y, label='Trajetória')
plt.axhline(mediay, color='green', linestyle='--', label='Valor Médio')
plt.xlabel('Tempo')
plt.ylabel('Posição em Y')
plt.title('Trajetória do bêbado em y')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(range(tempo_escolhido), dist_relativa, label='Distância Relativa')
plt.xlabel('Tempo')
plt.ylabel('Distância Relativa')
plt.title('Distancia relativa do bêbado')
plt.grid(True)
plt.show()
```

O gráfico inicial, relacionado com a posição em x pela posição em y do bêbado, corresponde ao ideal da teoria do andar do bêbado, seu comportamento é aleatório e imprevisível, cada vez que rodar a célula, ele irá se comportar de maneira diferente. Note que isso acontece pois para cada passo que o bêbado dá, o passo anterior nunca está relacionado com o próximo, visto que eles não são dependentes. Por conta disso, todas as decisões e caminhos sempre têm a mesma probabilidade de serem tomadas, e quando ele toma uma decisão, por exemplo, dar um passo para o norte, o próximo passo ainda terá as mesmas 4 possibilidades de caminho para serem tomadas (Norte, sul, leste e oeste).

Nos gráficos de trajetória em um eixo específico em relação ao tempo pode até ter alguma variância no comportamento, mas isso é justificado pela natureza estocástica, que é probabilística e indeterminada, podendo formar qualquer padrão estatístico, mesmo que sendo imprevisível, por exemplo com anomalias nas trajetórias dos eixos x e y. Esse padrão pode ser visto na cadeia de markov, que é um caso específico do processo estocástico, ela diz que os estados discretos (ou contínuos) de um sistema que tem a propriedade de distribuição da probabilidade de um próximo estado nada depende além de seu estado atual, e não da sequência de eventos precedidos aquele estado.

Já o último gráfico, referente à distância relativa do bebâdo em relação ao tempo, demonstra o comportamento de que o indivíduo aleatório tende a não voltar para a origem, visto que a probabilidade de ele estar em qualquer posição que não seja a origem é maior do que a probabilidade de ele estar na posição de origem no final de sua trajetória. Isso é comprovado pela sua própria trajetória e distância relativa, que muitas vezes não finaliza no ponto (0,0).
