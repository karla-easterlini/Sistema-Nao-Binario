```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

Definição do Teorema do Limite Central:
Se pegarmos um número grande de amostras da população (Com média e variância finitas), a distribuição da média da amostragem vai seguir uma distribuição normal, mesmo que a distribuição original não siga uma normal. 

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.array(Image.open(BytesIO(requests.get('https://media.geeksforgeeks.org/wp-content/uploads/20230529163945/Central-limit-theoram.png').content))))
plt.axis('off')
plt.show()
```

Quanto maior a amostragem, mais próxima será a aproximação da distribuição a uma Gaussiana. Como podemos ver na imagem:

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.array(Image.open(BytesIO(requests.get('https://www.statisticshowto.com/wp-content/uploads/2013/07/500px-Dice_sum_central_limit_theorem.svg_.png').content))))
plt.axis('off')
plt.show()
```

Esse gráfico mostra um exemplo de quantidade de dados de 6 lados rolados e seus possíveis valores a partir da distribuição usando o TLC.

Basicamente, o teorema diz que, para a maior quantidade de amostras possíveis, mais próximo o seu gráfico das médias será de uma gaussiana. 

Outra propriedade interessante, é que, pegando as médias das amostras, somando-as e tirando a média disso, equivalerá a um valor que será o mesmo que a média da população. O mesmo equivale para o desvio padrão, somando os desvios-padrão das amostras e calculando a média, será o desvio-padrão original da população. Essa propriedade pode ser útil em modelos de previsão para prever características específicas da população.

```python
#Os comentários na célula são apenas para observações, e não explicando os comandos.
from scipy.stats import norm
import statistics

n = [1, 5, 10, 50, 250, 1000, 5000, 20000, 100000] 
media = [] 
std = []
f = 0
z = []

for i in n:
    np.random.seed(15)
    arrs = [np.random.randint(-40, 41, i) for _ in range(1000)] #compreensão de lista
    z = np.mean(arrs, axis=1)
    st = np.std(arrs, axis=1)
    media.append(z)
    std.append(st)
    
k = 0

fig, ax = plt.subplots(3, 3, figsize=(9, 9))
for i in range(0, 3):
    for j in range(0, 3):
        # Fit a normal distribution to the data
        mu, std = norm.fit(media[k]) #Faz um fit da media, pra achar a média disto e o desvio padrão

        n, bins, patches = ax[i, j].hist(media[k], 50, density=True)

        # Plot the PDF using the fitted mean and standard deviation
        xmin, xmax = min(bins), max(bins)
        x = np.linspace(xmin, xmax, 250)
        p = norm.pdf(x, mu, std)
        ax[i, j].plot(x, p, 'k', linewidth=2)

        k = k + 1

plt.show()
```

Analisando o gráfico acima, é notável que, para uma maior quantidade de n (que varia de 1 a 1000), onde n define a quantidade de elementos dentro da matriz, mais próxima ela é de uma gaussiana e mais fina ela fica.

O interessante é, que o que a célula está calculando na verdade, são várias médias de amostras aleatórias distribuídas em um histograma (de área 1), demostrando um padrão normal. Quanto maior a quantidade de amostras, maior a quantidade de médias por amostra e maior será o comportamento do histograma com uma gaussiana. 

Outra observação interessante é que, para uma quantidade maior de médias distribuídas, menor é o desvio padrão encontrado.

Essa distribuição também pode ocorrer para por exemplo, uma distribuição uniforme que tem, para todos os valores, a mesma probabilidade de ocorrerem. Porém, mesmo sendo uniforme e no caso a seguir constante, essa distribuição, quando aplicada ao teorema do limite central, terá um comportamento normal:
